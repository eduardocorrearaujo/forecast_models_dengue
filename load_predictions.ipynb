```python
import requests 
```

```python
predictions_api = "https://api.mosqlimate.org/api/registry/predictions/"

page = 1
per_page = 50
pagination = f"?page={page}&per_page={per_page}&"

# List all Predictions
res = requests.get(predictions_api + pagination).json()

res
```

```python
# Filter by predict date
#requests.get(predictions_api + pagination).json()

# Filter using multiple parameters; predict date range
#requests.get(predictions_api + pagination).json()


# Advanced usage:
parameters = {
    "page": 1,
    "per_page": 50,
    "id": 5, 
    "model_id":3
    # Add parameters here
}

def get_predictions(parameters: dict):
    predictions_api = "https://api.mosqlimate.org/api/registry/predictions/?"
    parameters_url = "&".join([f"{p}={v}" for p,v in parameters.items()])
    return requests.get(predictions_api + parameters_url).json()

res = get_predictions(parameters)
```

```python
s = res['items'][0]['prediction']
```

```python
import pandas as pd
df = pd.DataFrame.from_records(s)
```

```python
import json 
json_struct = json.loads(s)    
df_flat = pd.io.json.json_normalize(json_struct)
df_flat.dates = pd.to_datetime(df_flat.dates)
df_flat.head()
```
